# K-means unsupervised learning

```python
#import the necessary libraries

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
from sklearn.cluster import KMeans
```

```python
#load iris dataset
X , y = datasets.load_iris(return_X_y=True)

X_t=X[:,0:4]
y_t=y[:]
```

```python
#Dimension of dataset
print("Dimension of X is : ", X.shape)
print("Dimension of y is : ", y.shape)
```

```python
#To store obtain result for different number ofcluster
wcss = []

for i in range(1, 16):
    #Set different hyperparameters for K-means
    kmeans = KMeans(n_clusters = i, init = 'k-means++', 
                    max_iter = 100, n_init = 10, random_state = 12)
    
    #fit the data set for different value of k in each epoch
    kmeans.fit(X)
    
    #store the result
    wcss.append(kmeans.inertia_)
```

```python
#Plot the graph for elbow method to find out optimum value of k

plt.plot(range(1, 16), wcss)
plt.title('Elbow method')
plt.xlabel('# of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#from above graph we can see that optimum value for K is 3

kmeans = KMeans(n_clusters = 3, init = 'k-means++',
                max_iter = 100, n_init = 10, random_state = 12)

#apply this to predict datapoints of cluster
y_pred = kmeans.fit_predict(X)
```

```python
#different colors for each cluster
color=['blue','green','cyan']

#label of each cluster
label1=['setosa','versicolour','virginica']

for r in range(3):
    #scatter plot of each cluster
    plt.scatter(X[y_pred == r, 0], X[y_pred == r, 1], 
            c =color[r], label = label1[r])
    
#scatter plot of centroids
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], 
            c = 'red', label = 'Centroids')

plt.legend()
plt.xlabel('X-label')
plt.ylabel('Y-label')
plt.show()
```
